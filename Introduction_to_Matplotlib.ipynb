```python
import matplotlib.pyplot as plt
import numpy as py

students_names=['Riya','Nancy','Caleb','Zora','Luna','Naomi','Ahmed','Gideon']
students_marks=[39,23,50,43,41,37,44,19]
marks_perc=[]
for x in students_marks:
  res=(x/50)*100
  marks_perc.append(res)
print(marks_perc)
```

```python
def marks_line_chart():
  plt.plot(students_names,students_marks,marker='o')
  plt.title("Student Marks")
  plt.xlabel("Students Label")
  plt.ylabel("Marks Label")
  plt.show()
marks_line_chart()
```

```python
def percentage_bar_chat():
  plt.bar(students_names,marks_perc)
  plt.title("Student percentage")
  plt.xlabel("Students Label")
  plt.ylabel("percentage Label")
  plt.show()
percentage_bar_chat()
```
